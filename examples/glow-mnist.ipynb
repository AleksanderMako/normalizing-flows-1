## Glow normalizing flows

This notebook provides a simple example of training Glow on MNIST.

Note that, for the default architecture settings, you *will* need a GPU in order to run this code.

As can be seen below, 100 epochs took roughly 2.5 hours on a Google Cloud NVIDIA V100 GPU.

For MNIST, one could also probably get away with a smaller model. This can be adjusted via the `num_layers` and `depth` arguments, as well as the `coupling_nn_ctor` argument, which builds the deep neural networks used in the coupling layers.

```python
import sys
sys.path += ['..']
```

```python
import tensorflow as tf
import tensorflow_probability as tfp
import numpy as np
from normalizing_flows.flows import Invert
from normalizing_flows.flows.glow import GlowFlow, coupling_nn_glow
from normalizing_flows.models import FlowLVM
from normalizing_flows.models.optimization import LinearWarmupSchedule
from tensorflow.keras.datasets import mnist
#tf.config.experimental_run_functions_eagerly(True)
```

```python
(X_train,_),_ = mnist.load_data()
batch_size = 100
X_train = np.expand_dims(X_train, axis=-1)
# resizing to 32x32 makes things a bit easier vis-a-vis the squeeze op
resize = lambda x: tf.image.resize(x, (32,32), method='bilinear')
X_train_ds = tf.data.Dataset.from_tensor_slices(X_train.astype(np.float32))
X_train_ds = X_train_ds.map(lambda x: x / 255. - 0.5) \
                       .map(resize) \
                       .shuffle(buffer_size=X_train.shape[0]) \
                       .batch(batch_size)
```

```python
glow = Invert(GlowFlow(num_layers=3, depth=16, coupling_nn_ctor=coupling_nn_glow(min_filters=32, max_filters=128, num_blocks=1)))
steps_per_epoch = X_train.shape[0] // batch_size
# warm-up for 10 epochs
learning_rate = LinearWarmupSchedule(2.0E-4, num_warmup_steps=10*steps_per_epoch)
model = FlowLVM(glow, input_shape=(None,32,32,1), num_bins=256,
                optimizer=tf.keras.optimizers.Adam(learning_rate=learning_rate))
```

```python
model.train(X_train_ds, steps_per_epoch, num_epochs=100)
```

```python
samples = model.sample(n=6)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(3*5,2*4))
for i in range(6):
    plt.subplot(2,3,i+1)
    plt.imshow(np.squeeze(255*(samples[i]+0.5)), cmap='gray')
plt.show()
```
